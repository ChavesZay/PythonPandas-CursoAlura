## Selecciones especificas
### Particiones
Seleccion de solo algunas columnas y filas, por lo cual debemos particionar

```python
df['c1'] #Para ver solo una columna
```

```python
(df[['c1','c2']]) #Seleccionar 2 o m√°s columnas
```

```python
type(df[['c1','c2']])
```

```python
df
```

```python
df[1:3] #Seleccionar por indices, es con la posicion
```

```python
df[1:][['c1','c3']] #Seleccion porindice y por columnas
```

```python
df.loc['13'] #Parametro nom de columnas (c1,c2,c3,c4,11,12,13,14)
```

```python
df.loc['11']
```

```python
df.loc[['13','12']]
```

```python
df
```

```python
df.loc['11','c2'] #Devuelve el valor de interseccion entre la columna y la fila
```
