### Inmuebles Residenciales

```python
import pandas as pd
```

```python
datos= pd.read_csv('alquiler.csv', sep=';')
datos.head(10)
```

```python
list(datos['Tipo'].drop_duplicates()) #Lista con los valores unicos de la columna 'Tipo'
```

```python
residencial= ['Habitación','Casa','Departamento','Casa en condominio','Casa comercial','Casa de villa'] #Locales Residenciales
```

```python
residencial
```

```python
datos.head(10)
```

## Filtrando datos
Buscando solo los locales que son residenciales, segun la lista generada

```python
seleccion_residenciales=datos['Tipo'].isin(residencial) #Metodo isin() para filtrar los datos que esten en la lista enviada por parametro y devuelve una serie con valores booleam
seleccion_residenciales
```

```python
datos_residencial=datos[seleccion_residenciales] #entre [] el criterio de seleccion
datos_residencial #Solo mostrara los locales residenciales
```

```python
list(datos_residencial['Tipo'].drop_duplicates())
```

## Filtado de números pares

```python
import pandas as pd

numeros = [i for i in range(11)]
letras = [chr(i + 65) for i in range(11)]
nome_coluna = ['N']

df = pd.DataFrame(data = numeros, index = letras, columns = nome_coluna)
```

```python
df
```

```python
seleccion= df['N'].isin([i for i in range(11) if i % 2 == 0])
df = df[seleccion]
df
```
