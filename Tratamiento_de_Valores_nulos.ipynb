## Tratamiento de Datos Faltantes (Valores nulos)

```python
import pandas as pd
```

```python
datos=pd.read_csv('alquiler_residencial.csv', sep=';')
```

## Identificar si hay valores nulos

```python
datos.isnull() #Devuelve True si el valor es  nulo
```

```python
datos.notnull() #Devuelve True si el valor no es  nulo
```

```python
datos.info() #Información del dataFrame
```

```python
datos[datos['Valor'].isnull()] #Devuelve las lineas que devuelven True, las lineas con valores nulos
```

```python
cantLineasNull=datos.shape[0]
cantLineasNull
```

### Tratamiento eliminando los Valores Nulos

```python
#Como el valor no se puede determinar, entonse se eliminaran las lineas
cantLineasNull=datos.shape[0]
datos.dropna(subset=['Valor'], inplace=True) #dropna()->Elimina todos los valores nulos, en subset se especifica la columna que contendra los valores nulos
```

```python
Lineas=datos.shape[0]
print(cantLineasNull-Lineas)
```

```python
datos[datos['Valor'].isnull()]
```

```python
datos.head(10)
```

### Tratamiento de datos- eliminando los nulos en 'Mantenimiento' si el tipo es Departamento

```python
datos[datos['Mantenimiento'].isnull()]
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
seleccion=(datos['Mantenimiento'].isnull())&(datos['Tipo']=='Departamento')
```

```python
A=datos.shape[0]
datos=datos[~seleccion] #Se usa '~' para traer lo contrario del criterio de seleccion
B=datos.shape[0]
A-B
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

### Tratamiento Valores Nulos_Rellenar o cambiar los valores null



```python
#datos=datos.fillna(0) Cambia los null de todas las columnas en 0
datos=datos.fillna({'Mantenimiento':0,'Impuesto':0}) #Entre un diccionario se especifica el nom de la columna y el valor a cambiar
```

```python
datos[datos['Mantenimiento'].isnull()].shape[0]
```

```python
datos[datos['Impuesto'].isnull()].shape[0]
```

### Reestructuración de Indice

```python
datos.index = range(datos.shape[0])
datos
```
