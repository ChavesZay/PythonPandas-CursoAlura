## Agrupando Valores

```python
grupo_distrito=datos.groupby('Distrito')
type(grupo_distrito)
```

```python
grupo_distrito.groups #Devuelve el nom del distrito y las posiciones donde se encuentra ese valor
```

```python
grupo_distrito
```

```python
for barrio,data in grupo_distrito:
  print('{}->{}'.format(barrio,data.Valor.mean()))
```

```python
grupo_distrito['Valor'].mean().round(2) #La misma informacion
```

```python
grupo_distrito[['Valor','Mantenimiento']].mean().round(2) #Doble [[]] para mas de una variable
```

## Estadísticas Descriptivas

```python
grupo_distrito['Valor'].describe().round(2) #Esto sirve para identificar datos extraños fuera de la curva
```

```python
grupo_distrito['Valor'].aggregate(['min','max']) #Para ver solos indicadores especificos, hacer un resumen del resumen
```

### Renombrar columnas

```python
grupo_distrito['Valor'].aggregate(['min','max']).rename(columns={'min':'Minimo','max':'Maximo'})
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.rc('figure',figsize=(20,20))
```

```python
fig= grupo_distrito['Valor'].mean().plot.bar(color='blue')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor Medio del alquiler por Distrito',{'fontsize':22})
```

```python
fig= grupo_distrito['Valor'].max().plot.bar(color='blue')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor Maximo del alquiler por Distrito',{'fontsize':22})
```
