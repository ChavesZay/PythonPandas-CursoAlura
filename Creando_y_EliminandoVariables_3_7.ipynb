## Creando Nuevas Variables

```python
import pandas as pd
```

```python
datos=pd.read_csv('alquiler_residencial.csv',sep=';')
```

```python
datos.head(10)
```

### Nuestro Jefe necesita en el DataFrame una nueva variable que indique el Valor Bruto

```python
datos['Valor Bruto']=(datos['Valor'])+(datos['Mantenimiento'])+(datos['Impuesto']) #Se crea la varible y se introduce la formula
datos
```

### Nuestro Jefe necesita en el DataFrame una nueva variable que indique el Valor por metro cuadrado

```python
datos['Valor m2']=(datos['Valor'])/(datos['Area'])
datos
```

### Redondear o disminuir decimales de una variable

```python
datos['Valor m2']=datos['Valor m2'].round(2)
datos.head(10)
```

### Nuestro Jefe necesita en el DataFrame una nueva variable que indique el Valor Bruto por metro cuadrado

```python
datos['Valor Bruto m2']=((datos['Valor Bruto'])/(datos['Area'])).round(2)
datos
```

### Nueva Variable en que todo que sea casa,Casa de Villa o Casa de codominio el resultado de la variable sea Casa y si no Departamento

```python
casa =['Casa','Casa en condominio','Casa de villa']
```

```python
datos['Tipo Agrupado']=datos['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Departamento')
datos
```

# Ejercicios Extras

```python

alumnos = pd.DataFrame({
'Nombre': ['Ary', 'Katia', 'Denis', 'Beto', 'Bruna', 'Dara', 'Carlos', 'Alice'],
'Sexo': ['M', 'F', 'M', 'M', 'F', 'F', 'M', 'F'],
'Edad': [15, 27, 56, 32, 42, 21, 19, 35],
'Notas': [7.5, 2.5, 5.0, 10, 8.2, 7, 6, 5.6]},
columns = ['Nombre', 'Edad', 'Sexo', 'Notas'])
```

```python
alumnos['Notas-Média(Notas)'] = alumnos['Notas'].apply(lambda x: x - alumnos['Notas'].mean())
alumnos.head()
```

```python
alumnos['Rango Edad'] = alumnos['Edad'].apply(lambda x: 'Menor que 20 años' if x < 20 else ('Entre 20 y 40 años' if (x >= 20 and x <= 40) else 'Mayor que 40 años'))
alumnos.head()
```

```python
alumnos['Notas-Média(Notas)'] = alumnos.Notas - alumnos.Notas.mean()
```

### Eliminar Variables

```python
datos_aux=pd.DataFrame(datos[['Tipo Agrupado','Valor Bruto','Valor m2', 'Valor Bruto m2']])
datos_aux.head(10)
```

```python
datos_aux.pop('Valor Bruto m2') # El método pop elimina la columna especificada, y es diferente al metodo del porq te muestar por pantalla lo que eliminaste
```

```python
datos_aux.head(10)
```

```python
datos.drop(['Valor Bruto','Valor Bruto m2'], axis=1, inplace=True) #Metodo para eliminar columnas, se coloca axis=1 para especificar que se van a  eliminar columnas
datos.head(10)
```
