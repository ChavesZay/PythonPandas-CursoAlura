```python
import pandas as pd
```

```python
data=[[1,2,3],[4,5,6],[7,8,9]]
data
```

```python
list('321') #Convierte cada caracter en un elemento de la lista
```

```python
df=pd.DataFrame(data,list('321'),list('ZYX')) #data,indices,nom_columnas
df
```

# Ordenar el DataFrame con el método sort

### Ordenar por indices

```python
df.sort_index(inplace=True) #Ordenar por indice, metodo sort para ordenar y parametro inplace para que se guarde el DataFrame el resultado
df
```

### Ordenar por columnas

```python
df.sort_index(inplace=True, axis=1) #Agregar como parametro axis=1 para ordenar por columnas
df
```

### Ordenar el contenido

```python
 #Ordenar los valores, el contenido del dataFrame
df.sort_values(by='X',inplace=True) #El parametro by= 'nombre_columna por la que se va a ordenar', con ese metodo los indices cambian de orden tambien
df
```

```python
df.sort_values(by='3',axis=1,inplace=True) #Para ordenar las columnas, por eso se pone axis=1
df
```

```python
df.sort_values(by=['X','Y'],inplace=True)
df
```
