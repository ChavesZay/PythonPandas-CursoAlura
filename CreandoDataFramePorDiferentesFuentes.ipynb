Cuando usamos la función read_html() en una URL o archivo HTML que contenga más de una tabla, cada una es convertida en un DataFrame y todas se almacenan en una lista de Python, es decir, la función read_html() devuelve una lista de DataFrames.

```python
df_html=pd.read_html('datos/datos_html_1.html')
df_html[0] # 0 porq lo transforma en lista
```

```python
resp = requests.get('https://federalreserve.gov/releases/h3/current/default.htm')
df_html2 = pd.read_html(resp.text)  # Devolver una lista de DataFrames
len(df_html2) #El sitio tiene 3 tablas
```
