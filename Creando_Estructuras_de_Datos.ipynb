```python
import pandas as pd
```

# Series

Creando Serie

```python
data=[1,2,3,4,5]
```

```python
serie=pd.Series(data)
serie
```

```python
index=['Linea'+ str(i) for i in range(5)] #Generando indices personalizados
index
```

```python
serie=pd.Series(data=data,index=index)
serie
```

```python
data={'Linea'+str(i): i+1 for i in range(5)}
data
```

```python
serie2=pd.Series(data)
serie2
```

```python
serie3=serie2+2 #Aumentar los elementos en 2 unidades
serie3
```

# Operaciones con Series
Si ambas series tienen el mismo indice, entonces podemos hacer operaciones aritmetricas con ellas

```python
serie4=serie2+serie3 #Como tienen el mismo indice, entonces se pueden realizar operaciones
serie4
```

# DataFrame

```python
datos = [[1, 2, 3], [4, 5, 6]]
index = 'X,Y'.split(',')
columns = list('CBA')[::-1]
df = pd.DataFrame(datos, index, columns)
df
```

```python
df1 = pd.DataFrame({'A': {'X': 1}, 'B': {'X': 2}})
df2 = pd.DataFrame({'C': {'X': 3}, 'D': {'X': 4}})
pd.concat([df1, df2])
```

```python
datos = [('A', 'B'), ('C', 'D')]
df = pd.DataFrame(datos, columns = ['L1', 'L2'],  index = ['C1', 'C2'])
df
```

```python
datos = {'A': {'X': 1, 'Y': 3}, 'B': {'X': 2, 'Y': 4}}
df = pd.DataFrame(datos)
df
```
