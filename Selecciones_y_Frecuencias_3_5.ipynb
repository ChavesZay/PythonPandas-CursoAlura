## Selecciones y Frecuencias
Frecuencia es el # de veces que se repite cierto fenómeno, o sea la cantidad de líneas que vamos a encontrar según un criterio específico



```python
import pandas as pd
```

```python
datos=pd.read_csv('alquiler_residencial.csv',sep=';')
datos.head(10)
```

## Mi necesita la frecuencia de algunos indicadores que estan contenidos en la ultima base de datos generada

### Seleccione solamente los inmuebles clasificados con tipo 'Departamento'

```python
seleccion=datos['Tipo'] =='Departamento'
seleccion
```

```python
n1=datos[seleccion]
n1
```

```python
frecuenciaDepartamento=datos[seleccion].shape[0]
frecuenciaDepartamento
```

### Seleccione solamente los inmuebles clasificados con tipo 'Casa', 'Casa en condominio' y 'Casa de Villa'

```python
seleccion=(datos['Tipo']=='Casa')|(datos['Tipo']== 'Casa en condominio') | (datos['Tipo']=='Casa de villa')
seleccion
```

```python
n2=datos[seleccion]
n2
```

```python
frecuenciaCasas=datos[seleccion].shape[0]
frecuenciaCasas
```

### Seleccione los inmuebles con área entre 60 y 100 metros cuadrados, incluyendo los limites

```python
seleccion=(datos['Area']>=60) & (datos['Area']<=100)
seleccion
```

```python
n3=datos[seleccion]
n3
```

```python
frecuenciaArea=datos[seleccion].shape[0]
frecuenciaArea
```

### Seleccione los inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2.000,00

```python
seleccion=(datos['Cuartos']>=4) & (datos['Valor']< 2000)
frecuenciaCuartosValor=datos[seleccion].shape[0]
frecuenciaCuartosValor
```

```python
print("Nº de inmuebles clasificados con tipo 'Departamento' -> {}".format(frecuenciaDepartamento))
print("Nº de inmuebles clasificados con tipos 'Casa', 'Casa en condominio' y 'Casa de villa' -> {}".format(frecuenciaCasas))
print("Nº de inmuebles con área entre 60 y 100 metros cuadrados, incluyendo los limites -> {}".format(frecuenciaArea))
print("Nº de inmuebles que tengan por lo menos 4 cuartos y alquiler menor que $2.000,00 -> {}".format(frecuenciaCuartosValor))
```
